```python
#Dividing the datas as test and train
```

```python
y = df["Type"].values
x = df.drop("Type",axis = 1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
#We are gonna use EarlyStopping and Drop out to solve overfitting problems.

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units = 30, activation="relu")) 
#Units are amount of the norons we are going to use.
#It is suggested that you'd better to set units,according to how much layers you have.
model.add(Dense(units = 15, activation="relu")) #This rule says that it makes sense to get any numbers between 30 and 1.
model.add(Dense(units = 15, activation="relu")) # To put 15 makes sense.
model.add(Dense(units = 1, activation="sigmoid")) 
model.compile(loss="binary_crossentropy",optimizer="adam") #used binary_crossentropy for classification.
```

```python
model.fit(x=x_train, y=y_train, epochs=700,validation_data=(x_test,y_test),verbose=1) 
#to exaggerate the epochs value may cause overfitting
```

```python
model.history.history
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot() #overfitting , we can not use another data in this model that we dont have 
#Launches too much within itself to be consistent 
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=15,activation = "relu"))
model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
modelLoss = pd.DataFrame(model.history.history)
```

```python
modelLoss.plot() #it is stopped in epoch 191 and seems better
```

```python
model = Sequential()

model.add(Dense(units=30,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation = "sigmoid"))

model.compile(loss="binary_crossentropy",optimizer = "adam")
```

```python
model.fit(x=x_train, y=y_train, epochs = 700, validation_data = (x_test,y_test), verbose = 1, callbacks=[earlyStopping])
```

```python
lossDf = pd.DataFrame(model.history.history)
```

```python
lossDf.plot()
```

```python
predictions = (model.predict(x_test) > 0.5).astype("int32")
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
